```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style = 'whitegrid')
```

```python
items = pd.read_pickle('items.pkl')
```

```python
items
```

**1. How many manufacturers produced items for each brand?** Looking at number of unqiue manufacturers for each brand.

```python
brand_uniq = items.groupby(by = 'brand').nunique()
```

```python
brand_uniq.head(6)
```

```python
brand_uniq[brand_uniq['manufacturer'] > 2]
```

```python
(brand_uniq['manufacturer'] == 2).sum(), (brand_uniq['manufacturer'] == 1).sum()
```

- Items with no brand are produced by 208 unique manufacturers, Brands 93 and 120 have four manufacturers each, 16 brands have two manufacturers each, and the remaining 256 brands have a single manufacturer. 
- Similar deductions can be made for `category1`,`2` and `3`.

**2. What are the brands with the highest and lowest customer ratings and retail prices?** Looking at mean `customerRating` and `recommendedRetailPrice` for each brand.

```python
brand_mean = items.groupby(by = 'brand').mean()
```

```python
brand_mean.head(6)
```

```python
brand_mean.sort_values(by = 'recommendedRetailPrice')
```

- The 5 most expensive and least expensive brands can be seen above.
- To look at the average ratings for each brand, unrated items will have to be excluded first.

```python
nonzero_ratings = items[items['customerRating'] != 0]
```

```python
nonzero_ratings_mean = nonzero_ratings.groupby(by = 'brand').mean()
```

```python
nonzero_ratings_mean.head(6)
```

```python
# sort by rating, then price
nonzero_ratings_mean.sort_values(by = ['customerRating','recommendedRetailPrice'])
```

```python
# sort by price, then rating
nonzero_ratings_mean.sort_values(by = ['recommendedRetailPrice', 'customerRating'])
```

**3. Which `categories` have the highest and lowest customer ratings and retail prices?** Looking at mean `customerRating` and `recommendedRetailPrice` for each `category1`,`2`, and `3`.

- **3.1. Looking at `category1`**, first for all items and then for items with non-zero ratings.

```python
items.groupby(by = 'category1').mean().sort_values(by = 'recommendedRetailPrice')
```

```python
nonzero_ratings.groupby(by = 'category1').mean().sort_values(by = 'customerRating')
```

- In `category1`, Labels 1, 8 and 7 (in order) are seen to have the highest customer ratings while the Labels 8, 3, 1 have the highest average retail price.

- **3.2. Looking at `category2`**

```python
# 5 lowest retail prices by category2
items.groupby(by = 'category2').mean().sort_values(by = 'recommendedRetailPrice').head()
```

```python
# 5 highest retail prices by category2
items.groupby(by = 'category2').mean().sort_values(by = 'recommendedRetailPrice').tail()
```

```python
# 5 lowest customer ratings by category2
nonzero_ratings.groupby(by = 'category2').mean().sort_values(by = 'customerRating').head()
```

```python
# 5 highest customer ratings by category2
nonzero_ratings.groupby(by = 'category2').mean().sort_values(by = 'customerRating').tail()
```

- **3.3. Looking at `category3`**, first for all items and then for items with non-zero ratings.

```python
items.groupby(by = 'category3').mean().sort_values(by = 'recommendedRetailPrice')
```

```python
nonzero_ratings.groupby(by = 'category3').mean().sort_values(by = 'customerRating')
```

- For `category3`, Labels 1, 8 and 6 are seen to have the highest ratings and labels 5, 8 and 1 have the highest retail prices.
